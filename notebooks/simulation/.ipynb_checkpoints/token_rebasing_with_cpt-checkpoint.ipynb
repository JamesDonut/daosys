```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
TYPE_TREASURY_TOKEN = 'TREASURY'
TYPE_DEBT_TOKEN = 'DEBT'
TYPE_COLLATERAL_TOKEN = 'COLLATERAL'
TYPE_DEPOSIT_TOKEN = 'DEPOSIT'

class Token():
    
    def __init__(self, name, native_rate = None):
        self.__tname = name
        self.__native_rate = native_rate
        self.__balance = 0.0
               
    def get_balance(self):
        return self.__balance
 
    def set_token(self, token):
        self.__tname = token.get_name()
        self.__balance = token.get_balance()
        self.__native_rate = token.get_rate()  

    def add_token(self, token):
        self.__balance = self.__balance + token.get_balance()
        self.__native_rate = token.get_rate()   

    def subtract_token(self, token):
        self.__balance = self.__balance - token.get_balance()
        self.__native_rate = token.get_rate()  

    def get_rate(self):
        return self.__native_rate    
    
    def get_name(self):
        return self.__tname       
    
    def set_balance(self, balance):
        self.__balance = balance
        
    def set_rate(self, native_rate):
        self.__native_rate = native_rate 
        
    def add_deposit(self, deposit):
        self.__balance = self.__balance + deposit

class RebaseToken(Token):  
    
    def __init__(self, name, native_rate = None):
        super().__init__(name, native_rate)
                  
    def rebase(self, rebase):
        balance = rebase + self.get_balance()
        self.set_balance(balance)  

class TradeToken(Token):  
    
    def __init__(self, name, native_rate = None):
        super().__init__(name, native_rate) 
        
    def get_balance(self):
        return self.__balance         
    
    def apply(self, token, rate = None):
        rate = rate if token.get_rate() == None else token.get_rate()
        self.__balance = rate*token.get_balance()      
               
class TokenArray():  
    def __init__(self, name = None, N = 0):
        self.__tokens = [] if N == 0 else [Token(name)] * N
     
    def append_token(self, token, native_rate=None):
        
        newToken = Token(token.get_name())
        newToken.set_token(token)
        newToken.set_rate(native_rate)  
        self.__tokens.append(newToken) 

    def get_balance(self, k):
        return self.__tokens[k].get_balance()         
        
    def set_balance(self, balance, k):
        self.__tokens[k].set_balance(balance)  
        
    def add_deposit(self, balance, k):
        current_balance = self.get_token_balance(k)
        self.__tokens[k].set_balance(current_balance+balance)         
        
    def set_token(self, token, k):
        self.__tokens[k] = token
    
    def get_token(self,k):
        return self.__tokens[k]      
    
    def get_array(self):
        return self.__tokens
    
    def set_array(self):
        self.__tokens = arr 
        
    def length(self):
        return len(self.__tokens)
        
class AggregateTokenArray(Token):
    
    def __init__(self, name, tokenArray = TokenArray()):
        super().__init__(name, None)
        self.__tokens = tokenArray
        self.__balance = 0
        
    def get_balance(self):
        return self.__balance        
                      
    def apply(self):
        for token in self.__tokens.get_array():
            token_balance = token.get_balance()
            self.__balance = self.__balance + token_balance

    
```

```python
N = 100

dToken = Token(TYPE_DEBT_TOKEN)
tToken = Token(TYPE_TREASURY_TOKEN)
tTokenArr = TokenArray(TYPE_TREASURY_TOKEN)

depositToken1 = Token(TYPE_DEPOSIT_TOKEN)
depositToken2 = Token(TYPE_DEPOSIT_TOKEN)
depositTokenArr = TokenArray(TYPE_DEPOSIT_TOKEN)

cToken1 = RebaseToken(TYPE_COLLATERAL_TOKEN)
cToken2 = RebaseToken(TYPE_COLLATERAL_TOKEN)
cTokenArr = TokenArray(TYPE_COLLATERAL_TOKEN)

dToken1 = RebaseToken(TYPE_DEBT_TOKEN)
dToken2 = RebaseToken(TYPE_DEBT_TOKEN)
dTokenArr = TokenArray(TYPE_DEBT_TOKEN)
tradeDebtTokenArr1 = TokenArray(TYPE_DEBT_TOKEN)
tradeDebtTokenArr2 = TokenArray(TYPE_DEBT_TOKEN)



for k in range(N):
    
    dTokenPrev = dToken
       
    # setup deposit probabilities
    deposit1 = deposit2 = 0.0
    select = np.random.choice(3, 1, p=[0.25, 0.25, 0.5])[0]
    
    if(select == 0): 
        # deposit first token only with 25% probability
        deposit1 = np.random.gamma(2, 10, 1)[0]
    elif (select == 1):
        # deposit second token only with 25% probability
        deposit2 = np.random.gamma(2, 10, 1)[0]
    else:
        # deposit both tokens with 50% probability
        deposit1 = np.random.gamma(2, 10, 1)[0]
        deposit2 = np.random.gamma(2, 10, 1)[0]
    
    depositToken1.set_balance(deposit1)
    depositToken2.set_balance(deposit2)
    
    nativeDepositToken1 = TradeToken(TYPE_DEBT_TOKEN)
    nativeDepositToken1.apply(depositToken1, 1)
    
    nativeDepositToken2 = TradeToken(TYPE_DEBT_TOKEN)
    nativeDepositToken2.apply(depositToken2, 1)    
    
    depositTokenPool = TokenArray(TYPE_COLLATERAL_TOKEN)
    depositTokenPool.append_token(nativeDepositToken1)
    depositTokenPool.append_token(nativeDepositToken2)
    depositToken = AggregateTokenArray(TYPE_TREASURY_TOKEN, depositTokenPool)
    depositToken.apply()
    
    # Collateral tokens
    cToken1.add_token(nativeDepositToken1)
    cToken2.add_token(nativeDepositToken2)

    cTokenPool = TokenArray(TYPE_COLLATERAL_TOKEN)
    cTokenPool.append_token(cToken1)
    cTokenPool.append_token(cToken2)
    
    cToken = AggregateTokenArray(TYPE_TREASURY_TOKEN, cTokenPool)
    cToken.apply()

    # Debt tokens
    dToken1.add_token(nativeDepositToken1)
    dToken2.add_token(nativeDepositToken2)

    dTokenPool = TokenArray(TYPE_DEBT_TOKEN)
    dTokenPool.append_token(dToken1)
    dTokenPool.append_token(dToken2)

    dToken = AggregateTokenArray(TYPE_DEBT_TOKEN, dTokenPool)
    dToken.apply()
    
    # Trade off deposit tokens for debt token
    tradeDebtToken1 = TradeToken(TYPE_DEBT_TOKEN)
    tradeDebtToken1.apply(depositToken1, 3)
    tradeDebtTokenArr1.append_token(tradeDebtToken1)
    
    tradeDebtToken2 = TradeToken(TYPE_DEBT_TOKEN)
    tradeDebtToken2.apply(depositToken2, 3)
    tradeDebtTokenArr2.append_token(tradeDebtToken2)
    
    depositTokenArr.append_token(depositToken) 
    cTokenArr.append_token(cToken)
    dTokenArr.append_token(dToken)
   
```

```python
tSamp = list(range(N))
deposits = [token.get_balance() for token in depositTokenArr.get_array()]
collateral = [token.get_balance() for token in cTokenArr.get_array()]

fig, ax = plt.subplots(figsize=(15, 6))
plt.plot(tSamp, collateral, label='collateral') 
plt.plot(tSamp, deposits, label='deposits')
plt.title("Cumulative treasury tokens (in native token)", fontsize=20)
plt.ylabel("Native token", fontsize=16)
plt.xlabel('Days', fontsize=16)
plt.legend()

trade1 = [token.get_balance() for token in tradeDebtTokenArr1.get_array()]
trade2 = [token.get_balance() for token in tradeDebtTokenArr2.get_array()]
cum_trade1 = list(np.cumsum(trade1))
cum_trade2 = list(np.cumsum(trade2))

fig, ax = plt.subplots(figsize=(15, 6))
plt.plot(tSamp, cum_trade1, label='Debt token trade (T1)') 
plt.plot(tSamp, cum_trade2, label='Debt token trade (T2)')
plt.title("Cumulative trades (in Debt token)", fontsize=20)
plt.ylabel("Debt token", fontsize=16)
plt.xlabel('Days', fontsize=16)
plt.legend()
```
